```python
IBM['Attrition'] = IBM['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
IBM['OverTime'] = IBM['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
IBM.Education.replace({1: 'High School',
                       2:'Undergrad',
                       3:'Graduate',
                       4:'Post Graduate',
                       5:'Doctorate'},inplace=True)
```

```python
columns = ["JobInvolvement", "JobSatisfaction", "PerformanceRating"]
```

```python
 IBM[columns].replace({1 : "Low",
                                    2 : "Medium",
                                    3 : "High",
                                    4 : "Very High"}, inplace = True)
```
